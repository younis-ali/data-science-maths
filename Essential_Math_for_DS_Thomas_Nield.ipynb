```python
# Limits

from sympy import *
from sympy.plotting import plot

x = symbols('x')
f = 1/x
result = limit(f, x, oo) # limit of the function 1/x such that x-> infinity
print(result)
```

```python
# Discovering eular number using limits
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n

euler_number = limit(f, n, oo)
print(f'Euler number is {euler_number.evalf()}')
```

# Descriptive and Inferential Statistics

## Descriptive Statistics

```python
# create a random list for experiments

import random

random.seed(0)
population = [random.randint(0, 100) for x in range(200)]
print(population)
```

```python
# Measures of Central Tendency using core pyton

def compute_mean(data):
    return sum(data)/len(data)

def compute_median(data):
    n = len(data)
    mid = int(n/2)-1 if n%2==0 else int(n/2) 
    ordered_data = sorted(data)
    if n % 2 == 0:
        return (ordered_data[mid] + ordered_data[mid + 1]) / 2.0
    else:
        return ordered_data[mid]

print(f"Mean = {compute_mean(population)}")
print(f"Median = {compute_median(population)}")
```

```python
# Measures of Central Tendency using pandas

import pandas as pd

# create the dataframe from the list
data = pd.DataFrame({"values" : population})

print(f"Mean = {data['values'].mean()}")
print(f"Median = {data['values'].median()}")
type(f"Mode = {data['values'].mode()}")
```

```python
# Ploting on data frame
data['values'].plot(kind = 'line', xlabel = "index", ylabel = "values", marker='o', linestyle='-', color='g')
```

```python
# Plotting usisng matplotlib
import matplotlib.pyplot as plt

# Plotting the line plot
plt.figure(figsize=(12, 6))
plt.plot(population, marker='o', linestyle='-', color='b')
plt.title('Line Plot of Population Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# Calcualte Variance and Standard Deviation using core python
from math import sqrt

def compute_sd(data):
    mean = sum(data)/len(data)
    variance = sum([(v - mean) ** 2 for v in data])/len(data)
    sd = sqrt(variance)
    print(f"Mean = {mean}")
    print(f"Variance = {variance}")
    print(f"Standard Deviation = {sd}")

compute_sd(population)
```

```python
# Calcualte Variance and Standard Deviation using pandas

mean = data['values'].mean()
variance = data['values'].var()
sd = data['values'].std()

print(f"Mean = {mean}")
print(f"Variance = {variance}")
print(f"Standard Deviation = {sd}")
```

```python
############ Normal distribution ###########

import numpy as np
import matplotlib.pyplot as plt

# Standard normal distribution

mean = 0
std = 1
num_samples = 100000

samples = np.random.normal(mean, std, num_samples)

# Plot the histogram of the samples
plt.hist(samples, bins=100, density=True, alpha=0.6, color='g')
plt.title("Normal Distribution")

plt.show()

```

```python
# Plot the PDF (probability density function)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi) * std)) * np.exp(-0.5 * ((x - mean) / std)**2)

plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mean = %.2f,  std dev = %.2f" % (mean, std)
plt.title(title)

plt.show()
```
