```python
from sympy import *

# SymPy variables are defined using symbols
x  = symbols('x')

# Plotting a linear function
f = 2*x + 1
plot(f)
```

```python
# Plotting a exponential funcion

from sympy import *

x = symbols('x')
f = x**2 + 1
plot(f)
```

```python
# Plotting 3D graphs with two independent variables and one dependent variable

from sympy import *
from sympy.plotting import plot3d

x, y =  symbols('x, y')
f = 2*x + 3*y
plot3d(f)
```

```python
# Limits

from sympy import *
from sympy.plotting import plot

x = symbols('x')
f = 1/x
result = limit(f, x, oo) # limit of the function 1/x such that x-> infinity
print(result)
```

```python
# Discovering eular number using limits
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n

euler_number = limit(f, n, oo)
print(f'Euler number is {euler_number.evalf()}')
```

```python
# Derivative calculator
from sympy import *

x = symbols('x')
f= x**2
dx_f = diff(f)
print(dx_f)
```
